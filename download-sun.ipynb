```python
import os
from dotenv import load_dotenv
from sunpy.net import attrs as a
from sunpy.net import Fido
from sunpy.time import TimeRange

load_dotenv()

START_TIME = "2011-02-13 18:20"
END_TIME = "2011-02-13 18:50"
EMAIL = os.environ['EMAIL']

def get_files(start_time, end_time, series, segment=None, email=None, save_dir="."):
    time_range = TimeRange(start_time, end_time)
    if segment:
        query = Fido.search(
            a.Time(time_range.start, time_range.end),
            a.jsoc.Series(series),
            a.jsoc.Segment(segment),
            a.jsoc.Notify(email)
        )
    else:
        query = Fido.search(
            a.Time(time_range.start, time_range.end),
            a.jsoc.Series(series),
            a.jsoc.Notify(email)
        )
    Fido.fetch(query, path=save_dir)


get_files(START_TIME, END_TIME, "hmi.B_720s", email=EMAIL, save_dir=r".\data\for_surya")
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
EMAIL = os.environ['EMAIL']


from sunpy.net import Fido
from sunpy.time import TimeRange
import sunpy.net.attrs as a
import astropy.units as u


def get_files(time_range, series, wavelength=None, segment=None, cadence=None, email=None, save_dir="."):
    time_range = TimeRange(time_range[0], time_range[1])

    attrs = [
        a.Time(time_range.start, time_range.end),
        a.jsoc.Series(series),
    ]

    if wavelength:
        attrs.append(a.Wavelength(wavelength * u.angstrom))
    if segment:
        attrs.append(a.jsoc.Segment(segment))
    if cadence:
        attrs.append(a.Sample(cadence * u.s))
    if email:
        attrs.append(a.jsoc.Notify(email))

    query = Fido.search(*attrs)
    files = Fido.fetch(query, path=save_dir)
    return files
```

```python
def get_all_files(time_ranges, email, save_dir):
    main_path = os.path.join(save_dir, "for_surya")
    os.makedirs(main_path, exist_ok=True)

    vectors_path = os.path.join(save_dir, "vectors")
    os.makedirs(vectors_path, exist_ok=True)

    for time_range in time_ranges:
        print(f"Downloading for time range: {time_range}")
        get_files(time_range, "aia.lev1_euv_12s", segment="image", cadence=720, email=email, save_dir=main_path)
        get_files(time_range, "aia.lev1_uv_24s", segment="image", cadence=720, email=email, save_dir=main_path) # 1600, 1700
        get_files(time_range, "hmi.B_720s", segment="field", email=email, save_dir=vectors_path)
        get_files(time_range, "hmi.B_720s", segment="azimuth", email=email, save_dir=vectors_path)
        get_files(time_range, "hmi.B_720s", segment="inclination", email=email, save_dir=vectors_path)
        get_files(time_range, "hmi.M_720s", email=email, save_dir=main_path)
        get_files(time_range, "hmi.ic_720s", email=email, save_dir=main_path)
```
